```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loan_data = pd.read_csv('D:/Python_e/loan_status_prediction/strat_train_set.csv')
loan_data.info()
```

```python
loan_data['Loan_Status'] = loan_data['Loan_Status'].apply(lambda x: 1 if x == 'Y' else 0)
print(loan_data['Loan_Status'])
```

```python
correlation_matrix = loan_data.corr()

print("Correlation Matrix:")
print(correlation_matrix)

import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
